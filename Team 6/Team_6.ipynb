```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('shopping_trends.csv')
```

```python
df
```

```python
df=df.set_index('Customer ID')
```

```python
df.drop_duplicates()
```

```python
df = df.drop(columns = ['Previous Purchases','Promo Code Used','Discount Applied','Subscription Status'] )
```

```python
df
```

```python
df.loc[df['Frequency of Purchases'] == 'Annually', 'Frequency of Purchases'] = 1
df.loc[df['Frequency of Purchases'] == 'Quarterly', 'Frequency of Purchases'] = 4
df.loc[df['Frequency of Purchases'] == 'Monthly', 'Frequency of Purchases'] = 12
df.loc[df['Frequency of Purchases'] == 'Weekly', 'Frequency of Purchases'] = 52
df.loc[df['Frequency of Purchases'] == 'Fortnightly', 'Frequency of Purchases'] = 26
df.loc[df['Frequency of Purchases'] == 'Bi-Weekly', 'Frequency of Purchases'] = 104
df.loc[df['Frequency of Purchases'] == 'Every 3 Months', 'Frequency of Purchases'] = 4
```

```python
df
```

```python
df['Age_Category_Interact'] = df['Age'].astype(str) + '_' + df['Category']
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Frequency of Purchases', data=df)
plt.title('Box Plot of Frequency of Purchases by Category')
plt.xlabel('Category')
plt.ylabel('Frequency of Purchases')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Purchase Amount (USD)', data=df)
plt.title('Box Plot of Purchase Amount by Category')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Frequency of Purchases'], color='skyblue')
plt.title('Bar Graph of Category vs. Frequency of Purchases')
plt.xlabel('Category')
plt.ylabel('Frequency of Purchases')
```

```python
sum_by_category = df.groupby('Category')['Frequency of Purchases'].sum().reset_index()
plt.bar(sum_by_category['Category'], sum_by_category['Frequency of Purchases'])
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Category VS Frequency of Purchase')
plt.show()
```

```python
df['Age Category'] = 0
df.loc[(df['Age'] >= 1) & (df['Age'] < 13), 'Age Category'] = 1
df.loc[(df['Age'] >= 13) & (df['Age'] < 20), 'Age Category'] = 2
df.loc[(df['Age'] >= 20) , 'Age Category'] = 3
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
male_data = df[df['Gender'] == 'Male']
female_data = df[df['Gender'] == 'Female']
average_m = male_data.groupby('Age Category')['Frequency of Purchases'].mean().reset_index()
average_f = female_data.groupby('Age Category')['Frequency of Purchases'].mean().reset_index()
bar_width = 0.35
plt.bar(average_m['Age Category'] - bar_width/2, average_m['Frequency of Purchases'], width=bar_width, label='Male')
plt.bar(average_f['Age Category'] + bar_width/2, average_f['Frequency of Purchases'], width=bar_width, label='Female')
plt.title('Average Frequency of Purchases by Gender and Age Category')
plt.xlabel('Age Category')
plt.ylabel('Average Frequency of Purchases')
plt.xticks([1, 2, 3], ['Children', 'Teenagers', 'Adults'])
plt.legend()
plt.show()
```
