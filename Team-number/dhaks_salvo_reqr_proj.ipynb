```python
#imported the given
import numpy as np
import pandas as pd
import math
```

```python
csv_file=pd.read_csv("imdb-movies.csv")
print(csv_file)
```

```python
#Removing few upcoming unused coloumn...
csv_file=csv_file.drop("vote_count",axis='columns')
csv_file=csv_file.drop("budget_adj",axis='columns')
csv_file=csv_file.drop("revenue_adj",axis='columns')
csv_file=csv_file.drop("production_companies",axis='columns')
csv_file=csv_file.drop("tagline",axis='columns')
csv_file=csv_file.drop("homepage",axis='columns')
csv_file=csv_file.drop("popularity",axis='columns')
csv_file=csv_file.drop("overview",axis='columns')
csv_file=csv_file.drop("id",axis='columns')
print(csv_file)
```

```python
csv_file=csv_file.dropna()
csv_file = csv_file[~(csv_file == 0).any(axis=1)]
print(csv_file)
```

```python
csv_file=csv_file.set_index("imdb_id")
print(csv_file)
```

<h1> Detect outliers in any column of your choice and plot them using a box plot.</H1>

<h3>For column runtime</h3>

```python
Q1_Y = csv_file["runtime"].quantile(0.25)
Q3_Y = csv_file["runtime"].quantile(0.75)
IQR_Y = Q3_Y-Q1_Y
csv_file[np.logical_or(csv_file["runtime"] < (Q1_Y-1.5 * IQR_Y), csv_file["runtime"] > (Q3_Y + 1.5 * IQR_Y))]
```

```python
csv_file.boxplot(column="runtime")
```

```python
#print movies with third-lowest budget
smallest3=csv_file.nsmallest(3,"budget")
lis=smallest3.index
print("The Details of movies with third-smallest budget")
print(smallest3.loc[lis[2]])
```

```python
#print movies with third-highest budget
largest3=csv_file.nlargest(3,"budget")
lis=largest3.index
print("The Details of movies with third-highest budget")
print(largest3.loc[lis[2]])
```

```python
#most and least earned revenue
print("most earnd revenue\n")
print(csv_file.nlargest(1,"revenue"))
print("\n\n\n\n\n")
print("least  earnd revenue\n")
print(csv_file.nsmallest(1,"revenue"))
```

```python
#10. What is the average number of words in movie titles between the years 2000-2005?

from numpy.lib.function_base import average
filtered_csv_file = csv_file[(csv_file["release_year"] <= 2005) & (csv_file["release_year"] >= 2000)]
count_list=[]
for data in filtered_csv_file["original_title"]:
  data=data.strip()
  data=data.split()
  count_list.append(len(data))
print(count_list)
print("Average title length: ",math.trunc(average(count_list)))
```
